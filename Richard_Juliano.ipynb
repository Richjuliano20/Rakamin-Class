```python
import pandas as pd
```

```python
path = "/online_retail_II (1).csv"
data = pd.read_csv(path, error_bad_lines=False)
data
```

```python
df = pd.DataFrame(data)
data['Year'] = pd.DatetimeIndex(df['InvoiceDate']).year
data
```

```python
df = df[df['Quantity'] >= 1]
# df[df.Quantity.ge(1)]
df
```

```python
# filter = data["Invoice"].isin(["C"])
df = df[~df["Invoice"].str.contains("C")]
df
```

```python
df['Revenue'] =  df['Quantity']*df['Price']
df
```

```python
df.groupby('Year')['Revenue'].mean()
```

```python
import numpy as np
# result_group_year= df.groupby(['Year'])
# total_by_year = result_group_year['Revenue'].agg([np.sum])

# total_by_year.plot(kind='bar' ,x='Year',y='sum',rot=0)
# # result_group_year
result_group_year= df.groupby(['Year'])
total_by_year = result_group_year['Revenue'].agg([np.mean]).reset_index()
total_by_year.plot(kind='bar' ,x='Year',y='mean',rot=0,  legend=False)
```
